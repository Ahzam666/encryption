```python
from PIL import Image
import numpy as np

def encrypt_image(input_image_path,encrypted):
    img = Image.open(input_image_path)
    
    img= np.array(img)
    img_array = img%16
    og = img-img_array
    og2 = og//16
    
    enc= Image.open(encrypted)
    enc = np.array(enc)
    img_array = enc%16
    enc2 = enc-img_array
    encrypted = enc+og2

    encrypted_img_array = encrypted  # Simple XOR encryption with key
    encrypted_img = Image.fromarray(encrypted_img_array)

    return encrypted_img

def decrypt_image(encrypted_image):
    encrypted_img_array = np.array(encrypted_image)
    decrypted_img = encrypted_img_array % 16
    decrypted_img_array = decrypted_img *16
    # Decrypt using the same key
    decrypted_img = Image.fromarray(decrypted_img_array.astype('uint8'))

    return decrypted_img

# Example usage
input_image_path = 'og.bmp'
encryption_key = 'ec.jpg'

encrypted_image = encrypt_image(input_image_path, encryption_key)
encrypted_image.save('encrypted_image.png')

decrypted_image = decrypt_image(encrypted_image, encryption_key)
decrypted_image.save('decrypted_image.png')
```

```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import numpy as np

def decrypt_image():
    encrypted_img = Image.open(encrypted_image_path)
    encrypted_img_array = np.array(encrypted_img)
    decrypted_img_array = encrypted_img_array % 16
    decrypted_img_array = decrypted_img_array * 16
    decrypted_img = Image.fromarray(decrypted_img_array.astype('uint8'))
    
    # Display the decrypted image
    decrypted_photo = ImageTk.PhotoImage(decrypted_img)
    decrypted_label.config(image=decrypted_photo)
    decrypted_label.photo = decrypted_photo  # Prevent PhotoImage from being garbage collected

def browse_input_image():
    global input_image_path
    input_image_path = filedialog.askopenfilename(filetypes=[("Image files", "*.bmp *.png *.jpg *.jpeg")])
    input_img = Image.open(input_image_path)
    input_photo = ImageTk.PhotoImage(input_img)
    input_label.config(image=input_photo)
    input_label.photo = input_photo  # Prevent PhotoImage from being garbage collected

def browse_encrypted_image():
    global encrypted_image_path
    encrypted_image_path = filedialog.askopenfilename(filetypes=[("Image files", "*.bmp *.png *.jpg *.jpeg")])
    encrypted_img = Image.open(encrypted_image_path)
    encrypted_photo = ImageTk.PhotoImage(encrypted_img)
    encrypted_label.config(image=encrypted_photo)
    encrypted_label.photo = encrypted_photo  # Prevent PhotoImage from being garbage collected

# Create the main application window
root = tk.Tk()
root.title("Image Decryption Tool")

# Browse buttons
input_button = tk.Button(root, text="Browse Input Image", command=browse_input_image)
input_button.pack()

encrypted_button = tk.Button(root, text="Browse Encrypted Image", command=browse_encrypted_image)
encrypted_button.pack()

decrypt_button = tk.Button(root, text="Decrypt Image", command=decrypt_image)
decrypt_button.pack()

# Labels to display images
input_label = tk.Label(root)
input_label.pack()

encrypted_label = tk.Label(root)
encrypted_label.pack()

decrypted_label = tk.Label(root)
decrypted_label.pack()

# Start the GUI event loop
root.mainloop()
```
